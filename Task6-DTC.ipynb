# Importing Libraries & Loading Iris Dataset :

```python
import sklearn.datasets as datasets
import pandas as pd
import matplotlib.pyplot as plt

# Loading the iris dataset
iris=datasets.load_iris()

# Forming the iris dataframe
X=pd.DataFrame(iris.data, columns=iris.feature_names)


y=iris.target
```

# Train-Test Split 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

# Define the Decision Tree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(X_train,y_train)

print('Decision Tree Classifer Created Successfully')
```

```python
y_predict = dtc.predict(X_test)
y_predict
```

```python
print("Score: "+str(dtc.score(X,y,sample_weight=None)*100)+" %")
```

# Constructing Decision Tree using matplotlib

```python
from sklearn import tree


a=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
b=['setosa','versicolor','virginica']

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (7,8), dpi = 500)

tree.plot_tree(dtc, feature_names = a, class_names = b, filled = True);
```
